```python
df["score_diff"] = df["h_pts"] - df["a_pts"]
```

```python
def convert_clock(clock):
    clock = clock.replace("PT", "")
    
    minutes = clock.split("M")[0]
    seconds = clock.split("M")[1].replace("S","")
    
    return int(minutes) + float(seconds)/60


df["time_remaining"] = df["clock"].apply(convert_clock)
```

```python
import torch

x_train = torch.tensor(x_train, dtype=torch.float32)
x_test = torch.tensor(x_test, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)
y_test = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

x_train = x_train.to(device)
x_test = x_test.to(device)
y_train = y_train.to(device)
y_test = y_test.to(device)
```

```python
import torch.nn as nn

class WinModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 32),
            nn.ReLU(),
            nn.Linear(32, 16),
            nn.ReLU(),
            nn.Linear(16, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.net(x)

model = WinModel().to(device)
```

```python
import torch.optim as optim

criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 10

for epoch in range(epochs):
    model.train()

    # Forward pass
    outputs = model(x_train)
    loss = criterion(outputs, y_train)

    # Backprop
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    # Evaluate accuracy on train set
    with torch.no_grad():
        preds = (outputs > 0.5).float()
        acc = (preds == y_train).float().mean()

    print(f"Epoch {epoch+1}/{epochs}, Loss: {loss.item():.4f}, Accuracy: {acc.item():.4f}")
```

```python
model.eval()

with torch.no_grad():
    test_outputs = model(x_test)
    test_preds = (test_outputs > 0.5).float()
    test_acc = (test_preds == y_test).float().mean()

print("Test Accuracy:", test_acc.item())
```
